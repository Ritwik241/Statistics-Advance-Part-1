#1. What is a random variable in probability theory?

 - In probability theory, a random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.



#2. What are the types of random variable?

  - There are two main types of random variables in probability theory:

1. Discrete Random Variable
Takes on finite or countably infinite values.

The values are distinct and separate.

Often represents counts or categories.

2. Continuous Random Variable
Takes on uncountably infinite values (usually from an interval of real numbers).

Can represent measurements like time, height, temperature, etc.

#9. What is the Poisson distribution and where is it applied?

 - The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, provided the events happen independently and at a constant average rate.

 Use Poisson distribution when:

Events happen at a constant rate.

Events occur independently.

You're counting how many times an event happens, not when.

The probability of two or more events in an infinitesimally small interval is zero.

 - When independent random variables are summed, their normalized sum tends toward a normal distribution â€” even if the original variables themselves are not normally distributed â€” provided the sample size is sufficiently large.


 Why is CLT Important?

Simplifies complex real-world distributions into something manageable.

#15. What is the application of Z statistics in hypothesis testing?

  - Z-statistics (or Z-scores) are widely used in hypothesis testing when dealing with normally distributed data or large samples. They help determine whether to accept or reject a null hypothesis by comparing observed data to what we would expect under the null.

#22. What is the concept of expected value in a probability distribution?

   - The expected value (EV), also called the mean of a probability distribution, is a measure of central tendency. It represents the long-run average outcome if an experiment is repeated many times.
